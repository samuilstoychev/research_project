```python
def extract_metrics(filename): 
    f = open(filename, 'r')
    lines = f.readlines() 
    gpu_usage = None
    gpu_mem = None
    cpu_total = None
    ram_init = None 
    ram_diff = None
    for line in lines: 
        if "=> Average precision" in line: 
            accuracy = float(line.split(" ")[-1])
        if "=> Total training time" in line: 
            training_time = float(line.split(" ")[-2])
        if "GPU BEFORE EVALUATION:" in line: 
            gpu_usage, gpu_mem = eval(line[22:])
        if "CPU AT THE END:" in line:
            cpu_total = eval(line.split(":")[1])
            cpu_total = cpu_total[0] + cpu_total[1]
        if "RAM BEFORE TRAINING:" in line: 
            ram_init = float(line.split(":")[-1])
        if "PEAK TRAINING RAM:" in line: 
            ram_diff = float(line.split(":")[-1]) - ram_init
    return (accuracy, training_time, gpu_usage, gpu_mem, cpu_total, ram_diff)
```

```python
def calculate_metrics(files): 
    metrics = dict()
    
    for file in files: 
        units = int(file[:-10]) 
        if units in metrics: 
            metrics[units].append(extract_metrics(file))
        else: 
            metrics[units] = [extract_metrics(file)] 
    
    return metrics
```

```python
metrics = calculate_metrics(a)
```

```python
usages = sorted([(x, sum([y[2] for y in metrics[x]])/3) for x in metrics])
```

```python
mems = sorted([(x, sum([y[3] for y in metrics[x]])/3) for x in metrics])
```

```python
accs = sorted([(x, sum([y[0] for y in metrics[x]])/3) for x in metrics])
```

```python
training_times = sorted([(x, sum([y[1] for y in metrics[x]])/3) for x in metrics])
```

```python
cpus = sorted([(x, sum([y[4] for y in metrics[x]])/3) for x in metrics])
```

### _Accuracy_ vs _GPU Memory Consumption_ Trade-Off

```python
fig = plt.figure()
fig.set_figheight(5)
fig.set_figwidth(7)
ax1 = fig.add_subplot(111)
ax1.plot(list(zip(*accs))[0], list(zip(*accs))[1], 'bo-')
ax1.set_ylabel('Accuracy', color='b')
for tl in ax1.get_yticklabels():
    tl.set_color('b')

ax2 = ax1.twinx()
ax2.plot(list(zip(*mems))[0], list(zip(*mems))[1], 'ro-')
ax2.set_ylabel('GPU memory consumption (MB)', color='r')
for tl in ax2.get_yticklabels():
    tl.set_color('r')
ax1.set_xlabel("Generator size ($G_{FC} = G_{OUT}$)")
# plt.savefig("/Users/samuilstoychev/Desktop/dissertation_figures/mnist_tuning_tradeoff.svg")
```

### _GPU consumption_ for different generator sizes

```python
plt.plot(list(zip(*usages))[0], list(zip(*usages))[1], "r")
plt.scatter(list(zip(*usages))[0], list(zip(*usages))[1], c="r")
plt.xlabel("Generator size")
plt.ylabel("Average GPU consumption (%)")
plt.show()
```

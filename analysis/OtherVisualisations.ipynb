```python
import os
import matplotlib
import matplotlib.pyplot as plt

font = {'size' : 16}

matplotlib.rc('font', **font)

matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
```

# Plotting Class Balance in CK+

```python
%cd ~/ckplus_cross_subject_cropped
```

```python
train_balance = [len(os.listdir("train/class_" + str(i))) for i in range(8)]
test_balance = [len(os.listdir("test/class_" + str(i))) for i in range(8)]
```

```python
EMOTION_LABELS = ["neutral", "anger", "contempt", "digust", "fear", "happy", "sadness", "surprise"]
```

```python
plt.figure(figsize=(4, 4), dpi=80)
plt.bar(EMOTION_LABELS, train_balance)
plt.xticks(rotation=45, ha="right")
plt.xlabel("Expression")
plt.ylabel("Number of images")
plt.savefig("/Users/samuilstoychev/Desktop/dissertation_figures/ckplus_train_dist.svg")
```

```python
plt.figure(figsize=(4, 4), dpi=80)
plt.bar(EMOTION_LABELS, test_balance)
plt.xticks(rotation=45, ha="right")
plt.xlabel("Expression")
plt.ylabel("Number of images")
plt.savefig("/Users/samuilstoychev/Desktop/dissertation_figures/ckplus_test_dist.svg")
```
